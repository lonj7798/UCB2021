```python
def double_eights(n):
    """Return true if n has two eights in a row.
    >>> double_eights(8)
    False
    >>> double_eights(88)
    True
    >>> double_eights(2882)
    True
    >>> double_eights(880088)
    True
    >>> double_eights(12345)
    False
    >>> double_eights(80808080)
    False
    """
    "*** YOUR CODE HERE ***"
    digits = str(n)
    print(digits)
    
    while(digits[:-1]):
        if (digits[-1] == '8') & (digits[-2] == '8'):
                return True
        
        digits = digits[:-1]
        
    return False
```

```python
double_eights(111111111111111154780957098214652177)
```

```python
def bake(cake, make):
    if cake == 0:
        cake = cake + 1
        print(cake)
    if cake == 1:
        print(make)
    else:
        return cake
    return make
```
